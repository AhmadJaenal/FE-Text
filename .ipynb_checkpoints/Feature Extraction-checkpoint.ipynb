```python
# Import plugin
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd
import re
import json
from openpyxl import Workbook
from openpyxl import load_workbook
```

```python
df = pd.read_excel('Review.xlsx', sheet_name="Data Review")
df.head(10)
```

```python
# Select Feature
feat_drop = ['Star Rating']
df = df.drop(columns=feat_drop)
df.head()
```

```python
# Filter
def filter_by_ascii(text, min_ascii=32, max_ascii=126):
    filtered_text = ''.join(char for char in text if ord(char) < 33 or (ord(char) > 64 and ord(char) < 91) or (ord(char) > 96 and ord(char) < 123) or ord(char) > 126)
    return filtered_text

df['Review'] = df['Review'].apply(lambda x: filter_by_ascii(x, 10))

df.head()
```

```python
# Check null value
df.isnull().sum()
```

```python
# Case Folding
df['Review'] = df['Review'].str.lower()
df.head()
```

```python
# Stopword
wb = Workbook()

stop_word = set(stopwords.words('indonesian'))
def stopword_text(text):
    factory = StopWordRemoverFactory()
    stopwords = factory.create_stop_word_remover()
    stop = stopwords.remove(text)
    return stop

df['Review'] = df['Review'].apply(stopword_text)
df.head()

excel_file_path = "Review.xlsx"
with pd.ExcelWriter(excel_file_path, mode='a', if_sheet_exists='new') as writer:
    df.to_excel(writer, sheet_name="Stopword", index=False)
```

```python
# Stemming
def stemming_text(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    result = stemmer.stem(text)
    return result
df['Review'] = df['Review'].apply(stemming_text)

excel_file_path = "Review.xlsx"
with pd.ExcelWriter(excel_file_path, mode='a', if_sheet_exists='new') as writer:
    df.to_excel(writer, sheet_name="Stemming", index=False)
```

```python
clean_stemming = pd.read_excel('Review.xlsx', sheet_name="Clean Stemming")
file_path = 'combined_slangword.json'
with open(file_path, 'r') as file: kamus = json.load(file)
    
def check_slang(data, kamus):
    data_review = data['Review']
    for index, sentence in enumerate(data_review):
        new_sentence = " ".join(kamus.get(word, word) for word in sentence.split(' '))
        data['Review'][index] = new_sentence
    return data
            
result = check_slang(clean_stemming, kamus)

result
excel_file_path = "Review.xlsx"
with pd.ExcelWriter(excel_file_path, mode='a', if_sheet_exists='new') as writer:
    result.to_excel(writer, sheet_name="Result chek Slang Word", index=False)
```

```python
# Tokenize
data_review = pd.read_excel('Review.xlsx', sheet_name="Result chek Slang Word")

def tokenize(text):
    text = str(text)
    tokens = word_tokenize(text)
    return tokens

data_review['Review'] = data_review['Review'].apply(tokenize)

excel_file_path = "Review.xlsx"
with pd.ExcelWriter(excel_file_path, mode='a', if_sheet_exists='new') as writer:
    data_review.to_excel(writer, sheet_name="Tokenize", index=False)
```

```python
# Category
kata_bahan = [
    "bahan", "kain", "tekstur", "lembut", "nyaman",
    "kokoh", "awet", "berat", "ringan", "elastis",
    "transparan", "halus", "kasar", "berbulu", "tipis",
    "tebal", "lentur", "berkerut", "pudar", "mudah robek",
    "rapi", "adem", "hangat", "kuat", "fleksibel", "kering",
    "mengkilap", "matte", "anti air", "berpori", "licin"
]

kata_model = [
    "desain", "model", "motif", "potongan", "siluet",
    "warna", "garis", "detail", "jahitan", "kerapian",
    "elegan", "modern", "klasik", "trendi", "minimalis",
    "berlebihan", "unik", "kreatif", "simpel", "rumit",
    "modis", "fashionable", "gaya", "cantik", "rapi", "futuristik",
    "sablon", "bordir", "aplikasi", "patch", "printing", "grafis", "ukur"
]

def stemming_text_category(list_word):
    stemmer = PorterStemmer()
    stemmed_word = [stemmer.stem(word) for word in list_word]
    return stemmed_word

stemming_bahan = stemming_text_category(kata_bahan)
stemming_desain = stemming_text_category(kata_model)
```

```python
df_tokenize = data_review = pd.read_excel('Review.xlsx', sheet_name="Tokenize")
df_tokenize['Category'] = ''

for i in range(len(df_tokenize)):
    text_review = df_tokenize.iloc[i]['Review']
    if any(word in text_review for word in stemming_bahan):
       df_tokenize.loc[i, 'Category'] = 'Bahan'
    elif any(word in text_review for word in stemming_desain):
        df_tokenize.loc[i, 'Category'] = 'Model/Desain'
    else:
        df_tokenize.loc[i, 'Category'] = 'Tidak terkategorikan'
        
df_tokenize.head()

excel_file_path = "Review.xlsx"
with pd.ExcelWriter(excel_file_path, mode='a', if_sheet_exists='new') as writer:
    df_tokenize.to_excel(writer, sheet_name="Aspek Base", index=False)   
```

```python
data_review = pd.read_excel('Review.xlsx', sheet_name="Aspek Base")
corpus = data_review['Review']
corpus
```

```python
labels = data_review['Sentimen'].tolist()
labels
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer 
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline 
from sklearn.metrics import classification_report, accuracy_score
```

```python
tfidf = TfidfVectorizer()
vectorizer = tfidf.fit_transform(corpus)
```

```python
pd.DataFrame(vectorizer.toarray(), columns=tfidf.get_feature_names_out())
```

```python
count_vector = CountVectorizer()
```

```python
vectorizer = count_vector.fit_transform(corpus)
```

```python
pd.DataFrame(vectorizer.toarray(), columns=count_vector.get_feature_names_out())
```

```python
def test():
    for i in range(100):
        X_train, X_test, y_train, y_test = train_test_split(corpus, labels, test_size=0.3, random_state=i)
        X_train_tfidf = tfidf.fit_transform(X_train)
        X_test_tfidf = tfidf.transform(X_test)
        model = MultinomialNB()
        model.fit(X_train_tfidf, y_train)
        y_pred = model.predict(X_test_tfidf)
        if(accuracy_score(y_test, y_pred) >= .94):
            print("Accuracy:", accuracy_score(y_test, y_pred))
            print("Classification Report:\n", classification_report(y_test, y_pred))

test()
```

```python
import seaborn as sns
sns.set_theme()

labels = ['Positif', 'Negatif']
counts = [total_review_positif, total_review_negatif]


def show_bar_chart(labels, counts, title):
    fig, ax = plt.subplots(figsize=(8, 6))

    bars = ax.bar(labels, counts, color=['#2394f7', '#f72323', '#fac343'])

    for bar, count in zip(bars, counts):
        height = bar.get_height()

        ax.annotate(f'{count}', xy=(bar.get_x() + bar.get_width() / 2, height),  xytext=(0, 3),

                                   textcoords="offset points",
                                   ha='center', va='bottom')

    ax.grid(axis='y', linestyle='', alpha=0.7)

    ax.set_xlabel('Sentimen')
    ax.set_ylabel('Jumlah')
    ax.set_title(title)

    plt.show()

show_bar_chart(labels, counts, "Distribusi Sentimen Hotel")
```
